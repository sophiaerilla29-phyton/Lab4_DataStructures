```python
# Required Inputs

SEED_NUM = 5
STUDENT_MAJOR = "BSME"
FAVORITE_ARTIST = "NCT"
LOCAL_HAZARD = "TYPHOON"

CONTROL_NUM = max(3, SEED_NUM)

print("CONTROL_NUM:", CONTROL_NUM)
print("STUDENT_MAJOR:", STUDENT_MAJOR)
print("FAVORITE_ARTIST:", FAVORITE_ARTIST)
print("LOCAL_HAZARD:", LOCAL_HAZARD)
```

```python
from collections import defaultdict, Counter

stream = [
    ("TUP Manila", "BSECE"),
    ("TUP Taguig", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", "BSECE"),
    ("TUP Visayas", STUDENT_MAJOR),
    ("TUP Taguig", "BSECE"),
    ("TUP Manila", "BSECE")
]

custom_applications = [
    ("TUP Manila", STUDENT_MAJOR),
    ("TUP Taguig", STUDENT_MAJOR),
    ("TUP Visayas", f"BSCE_v{CONTROL_NUM}"),
    ("TUP Manila", "BSECE"),
    ("TUP Manila", f"BSME_v{CONTROL_NUM}"),
    ("TUP Taguig", "BSIT"),
    ("TUP Visayas", "BSECE")
]

stream.extend(custom_applications)

campus_applications = defaultdict(list)

for campus, program in stream:
    campus_applications[campus].append(program)

campus_counters = {}

for campus, programs in campus_applications.items():
    campus_counters[campus] = Counter(programs)

top_program, top_frequency = campus_counters["TUP Manila"].most_common(1)[0]

print("CONTROL_NUM Used:", CONTROL_NUM)
print("STUDENT_MAJOR Used:", STUDENT_MAJOR)
print("Expanded Stream Length:", len(stream))
print("Total Applications for TUP Manila:", len(campus_applications["TUP Manila"]))
print("Top Requested Program at TUP Manila:", top_program)
print("Exact Frequency of Top Program:", top_frequency)
```

```python
festival = {
    "Ben&Ben",
    "SB19",
    "Bini",
    "Eraserheads",
    FAVORITE_ARTIST,
    "Zild",
    f"Indie Artist {CONTROL_NUM}"
}

user_a = {"Ben&Ben", "Bini", "Maki", "Dionela", FAVORITE_ARTIST}
user_b = {"SB19", "Eraserheads", "Zild", f"Indie Artist {CONTROL_NUM}", "Parokya ni Edgar"}

festival.update({"Moira", "December Avenue", "Cup of Joe", "IV of Spades"})

intersection_a = user_a.intersection(festival)
union_a = user_a.union(festival)
jaccard_a = (len(intersection_a) / len(union_a)) * 100

intersection_b = user_b.intersection(festival)
union_b = user_b.union(festival)
jaccard_b = (len(intersection_b) / len(union_b)) * 100

dealbreakers_a = user_a.difference(festival)

print("Total Artists in Expanded Festival:", len(festival))
print("Length of Intersection User A & Festival:", len(intersection_a))
print("Length of Union User A & Festival:", len(union_a))
print("User A Jaccard Similarity (%):", round(jaccard_a, 2))
print("User B Jaccard Similarity (%):", round(jaccard_b, 2))
print("User A Dealbreaker Artists:", dealbreakers_a)
```

```python
from collections import deque, Counter

alert_queue = deque(maxlen=5)

burst = [
    (CONTROL_NUM, "WEATHER"),
    (CONTROL_NUM + 1, "TRAFFIC"),
    (CONTROL_NUM + 2, "WEATHER"),
    (CONTROL_NUM + 3, "WEATHER"),
    (CONTROL_NUM + 4, LOCAL_HAZARD),
    (CONTROL_NUM + 5, "WEATHER"),
    (CONTROL_NUM + 6, "TRAFFIC")
]

custom_alerts = [
    (CONTROL_NUM + 7, "POWER"),
    (CONTROL_NUM + 8, "TRAFFIC"),
    (CONTROL_NUM + 9, LOCAL_HAZARD),
    (CONTROL_NUM + 10, "WEATHER"),
    (CONTROL_NUM + 11, "FIRE"),
    (CONTROL_NUM + 12, "POWER")
]

burst.extend(custom_alerts)

execution_log = []

for alert in burst:
    timestamp, category = alert

    current_categories = [item[1] for item in alert_queue]
    category_count = Counter(current_categories)

    if category_count[category] >= 2:
        execution_log.append((alert, "REJECTED"))
    else:
        alert_queue.append(alert)
        execution_log.append((alert, "APPENDED"))

final_sum = sum(item[0] for item in alert_queue)

print("LOCAL_HAZARD Used:", LOCAL_HAZARD)
print("Total Burst Size:", len(burst))
print('Did "WEATHER" at Timestamp CONTROL_NUM + 3 append?: N')
print("Did LOCAL_HAZARD append?: Y")
print("Final Deque Length:", len(alert_queue))
print("Final Sum of Timestamps in Deque:", final_sum)
print("Exact Final Deque Output:", alert_queue)

print("\nExecution Log:")
for item in execution_log:
    print(item)
```
